# Zastosowania DFT

Na poczatku zajmiemy się wyszukiwaniem wzorca 'e' w tekscie za pomoca DFT.
Wczytujemy obrazek, odwracamy kolory, to samo robimy dla wzorca, nastepnie za pomocą DFT obliczamy korelacje miedzy wzrocem a obrazkiem.

```python
from PIL import Image
import matplotlib.pyplot as plt
import numpy as np
from numpy import fft
import cv2


galia = 255 - np.asarray(Image.open("lab10_galia.png").convert("L"))

plt.figure(figsize=(15,8))
plt.title("Wczytany obrazek w skali szarości z odwróconymi kolorami",fontsize = '20')
plt.imshow(galia,cmap="gray")
plt.show()
```

```python
galia_fft = fft.fft2(galia)
plt.imshow(np.log10(np.abs(galia_fft)).astype(np.float64))
plt.title("Modul z wyniku DFT na naszym obrazku (w skali logarytmicznej)")
plt.show()
plt.title("Wartosci fazy z wyniku DFT na naszym obrazku")
plt.imshow(np.angle(galia_fft))
plt.show()
```

```python
galia_e = 255 - np.asarray(Image.open("lab10_galia_e.png").convert("L"))
plt.imshow(galia_e,cmap= "gray")
plt.title("Wzorzec ktory bedziemy wyszukiwac")
plt.show()
```

```python
corr = np.real(
    np.fft.ifft2(
        np.multiply(
            fft.fft2(galia),
            fft.fft2(np.rot90(galia_e,2),(galia.shape[0],galia.shape[1]))
            )
        )
    )

result = np.array(Image.open("lab10_galia.png"))

max_corr = np.amax(corr)

for i in range(corr.shape[0]):
    for j in range(corr.shape[1]):
        if corr[i,j] > 0.95 * max_corr:
            result[i-galia_e.shape[0]:i,j-galia_e.shape[1]:j] = (255,255,0)
            
plt.figure(figsize=(15,8))
plt.title("Wejsciowy obrazek z zaznaczonymi miejscami wystapienia wzorca",fontsize=20)
plt.imshow(result,cmap = "gray")
plt.show()
```

Jest w stanie obsluzyc jeden typ i rozmiar czcionki. Obsluguje tekst skladajacy sie z wielu linii (wykrywa je).
Tekst na obrazku moze byc obrócony.

```python
from PIL import Image
import matplotlib.pyplot as plt
import numpy as np
from numpy import fft
import cv2
from PIL import ImageFont
from PIL import ImageDraw

img = np.asarray(Image.open("ocr_text.png").convert("L"))
plt.figure(figsize=(15,8))
plt.imshow(img,cmap="gray")
plt.title("Wczytany obrazek wejsciowy",fontsize=20)
plt.show()
```

Funkcja obacajaca i wyrownujaca obrazek, poprzez znalezienie minimalnego prostokata w ktorym znajduje sie tekst.

```python
def rotate_img(img):
    
    box = np.column_stack(np.where(img > 0))
    angle = cv2.minAreaRect(box)[2]

    angle = -angle
    if angle > 45:
        angle -=90

    rotation_matrix = cv2.getRotationMatrix2D([x//2 for x in img.shape], angle, 1.0)
    rotated = cv2.warpAffine(img, rotation_matrix, img.shape[::-1])
    return rotated
```

Tutaj nastepuje jeszcze odwrocenie kolorów obrazka

```python
img = 255 - img
img = rotate_img(img)
plt.figure(figsize=(15,8))
plt.imshow(img,cmap="gray")
plt.title("Obrazek z wyrownanym tekstem i odwroconymi kolorami",fontsize=20)
plt.show()
```

Definiujemy sobie nasz alfabet skadajacy sie z malych liter alfabetu lacinskiego, cyfr oraz znakow przestankowych.
Tworzymy sobie takze slownik w ktorym przechowujemy kazda z liter z naszego alfabetu oraz jej reprezentacje macierzowa, ktora przyda nam sie pozniej w dopasowywaniu wzorca.

```python
chars = {}
alphabet =['a','b','c','d','e','f','g','h','i','j','k','l','m','n','o','p','r','s','t','u','w','q','v','x','y','z','1','2'
            ,'3','4','5','6','7','8','9','.','!','?',','] 

font = ImageFont.truetype("Arial.ttf",size = 60)
for char in alphabet:
    letter = Image.new('L',font.getsize(char),255)
    draw = ImageDraw.Draw(letter)
    draw.text((0,0),char,font = font)
    chars[char] = 255 - np.array(letter)
    
```

Znajdujemy bounding boxy wszystkich liter na obrazku bardzo przydatne w tym sa dwie funkcje z biblioteki opencv
cv2.findContours() oraz cv2.boundingRect(). Warto zauwazyc ze kropki w literach 'i' oraz 'j' dostaja osobny bounding box co stanowi pewien problem z ktorym potem trzeba bedzie sobie poradzic

```python
countours, _  = cv2.findContours(img,mode = cv2.RETR_EXTERNAL,method = cv2.CHAIN_APPROX_SIMPLE)
boxes_img = np.zeros((img.shape[0], img.shape[1], 3), dtype = np.uint8)
bounding_boxes = []
for c in countours:
    rect = cv2.boundingRect(c)
    bounding_boxes.append(rect)
    cv2.rectangle(boxes_img,(rect[0],rect[1]),(rect[0]+rect[2],rect[1]+rect[3]),(255,255,0))

plt.figure(figsize = (15,8))
plt.imshow(boxes_img)
plt.title("Znalezione bounding boxy",fontsize=20)
plt.show()
```

Zdefiniujmy sobie funkcje pozwalajaca sklasyfikowac znak znajdujacy sie na obrazku, dla kazdej litery obliczana jest korelacja za pomoca DFT oraz nastepnie wybierana jest litera z najwieksza korelacja.

```python
def match_letter(box_img, chars):
    
    xSize = max(max([img.shape[1] for img in list(chars.values())]),box_img.shape[1])
    ySize = max(max([img.shape[0] for img in list(chars.values())]),box_img.shape[0])
    
    image = np.zeros((ySize,xSize))
    image[:box_img.shape[0],:box_img.shape[1]] = box_img
    
    max_char = None
    max_corr = -float("inf")
    
    for char, char_img in chars.items():
        pattern= np.zeros_like(image)
        pattern[:char_img.shape[0], :char_img.shape[1]] = char_img
        corr = np.real(
        np.fft.ifft2(
            np.multiply(
                np.fft.fft2(image-100),
                np.fft.fft2(np.rot90(pattern-100, 2))
                    )
                )
            )
        curr_corr = np.amax(corr)
        if curr_corr > max_corr:
            max_char = char
            max_corr = curr_corr
            
    return max_char
```

Dla kazdego bounding boxa klasyfikujemy znak znajdujacy sie w nim, tworzymy sobie liste w ktorej przechowujemy wszystkie bounding boxy wraz z przyporzadkowanymi im znakami.

```python
letters = []
for box in bounding_boxes:
    box_img = img[box[1]:box[1]+box[3],box[0]:box[0]+box[2]]
    letters.append((box,match_letter(box_img,chars)))
```

Nastepnie wyszukujemy linie na naszym obrazku sumujac wszystkie kolumny obrazka do jednej co pozwala nam na znalezienie linii poprzez sprawdzenie ciaglosci w zsumowanej kolumnie.

```python
def find_lines(image):
    summed = np.sum(image,axis = 1)
    in_line = False
    lines = []
    line_idx = 0
    for i,val in enumerate(summed):
        if val != 0 and not in_line:
            line_beggining = i
            in_line = True
        if val == 0 and in_line:
            in_line = False
            lines.append((line_beggining,i,line_idx))
            line_idx+=1
    return lines
```

```python
lines = find_lines(img)
```

Nastepnie musimy okreslic kolejnosc bounding boxow, czyli poprostu sortujemy sobie bounding boxy przyporzadkowujac je do linii w ktorych sie znajduja a nastepnie sortujac je w obrebie jednej linii. Warto dodac ze na tym etapie bedziemy odrzucac kropki z literek 'i' oraz 'j' ktore zostaly sklasyfikowane jako znak przestankowy.

```python
lines_y_sorted = [[] for _ in range(len(lines))]

for line in lines:
    for letter in letters:
        middle = letter[0][1] + letter[0][3]/2
        if line[0] < middle and line[1] > middle:
            lines_y_sorted[line[2]].append(letter)

lines_sorted = [sorted(line,key = lambda x: x[0][0]) for line in lines_y_sorted]

#stip dots
out_lines = []
for line in lines_sorted:
    stripped_line = []
    for i in range(len(line)):
        if line[i][1] == '.' and i > 0 and (line[i-1][1] == 'i' or line[i-1][1] == 'j'):
            continue
        else:
            stripped_line.append(line[i])
    out_lines.append(stripped_line)
```

Znajdujemy najdluzsza przerwe pomiedzy znakami i nastepnie na jej podstawie wyznaczamy dlugosc spacji pomiedzy znakami wartosc longest_space * 0.5 zostala wyznaczona metoda prob i bledow

```python
longest_space = 0

for line in out_lines:
    for i in range(1,len(line)):
        longest_space = max(line[i][0][0] - (line[i-1][0][0]+line[i-1][0][2]),longest_space)

space_length = longest_space * 0.5
```

Tworzymy nasz wynikowy tekst

```python
result = ""
for line in out_lines:
    for i in range(len(line)):    
        if i > 0 and line[i][0][0] - (line[i-1][0][0]+line[i-1][0][2]) >= space_length:
            result+=" "
        result+=line[i][1]        
    result+="\n"
    
print(result)
    
```

Obliczamy odległość levenshteina naszego wyniku oraz oryginalnego tekstu na obazku

```python
original = "duis at turpis ut velit imperdiet sagittis sed tristique risus. mauris pulvinar tincidunt velit, id mollis nisl. duis vitae elementum mauris. nulla feugiat auctor nisi, sit amet iaculis ipsum posuere eget. nulla et urna venenatis, auctor justo vitae, volutpat diam. nullam quis dolor non odio dapibus convallis."
```

```python
def edit_distance(str1, str2):
    arr = [[(0) for _ in range(len(str2)+1)] for _ in range(len(str1)+1)]

    for i in range(len(str2)+1):
        arr[0][i] = (i)
    for i in range(len(str1)+1):
        arr[i][0] = (i)
    
    for row in range(1,len(str1)+1):
        for col in range(1,len(str2)+1):
            if str1[row - 1] == str2[col -1]:
                arr[row][col] = arr[row-1][col-1]
            else:
                min_cost = min(arr[row][col -1],arr[row-1][col],arr[row-1][col-1])
                if min_cost == arr[row][col -1]:
                    arr[row][col] = (arr[row][col -1]+1)
                elif min_cost == arr[row-1][col]:
                    arr[row][col] = (arr[row-1][col]+1)
                else:
                    arr[row][col] = (arr[row-1][col-1]+1)
    return arr[len(str1)][len(str2)]
```

```python
print(f"Odległość levenshteina: {edit_distance(result,original)}")
```

```python
from collections import Counter
result_count = Counter(''.join(filter(str.isalpha, result)))
original_count = Counter(''.join(filter(str.isalpha, original)))
print("Liczba wystapien liter w tekscie")
for key,val in original_count.items():
    print(f"Litera {key} tekst oryginalny: {val} tekst z OCR: {result_count.get(key,0)}")
```
